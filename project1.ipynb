```python
# CONSOLE-BASED SMART CALCULATOR
import numpy as np


def get_numbers():
  n=int(input("How many numbers do you want to enter? : "))
  num=[]

  for i in range(n):
      x=int(input("Enter the number: "))
      num.append(x)
  return num

# Addition
def Addition():
  num=get_numbers()

  npnum=np.sum(num)
  print(f"The addition of  numbers are : {npnum}")


# Subtraction
def Subtraction():
  num=get_numbers()

  result=num[0]
  for i in range(1,len(num)):
    result=np.subtract(result,num[i])
  sub=result
  print(f"The subtraction of  numbers are :{sub}")

# Multiplication
def Product():
  num=get_numbers()

  npnum=np.prod(num)
  print(f"The product of  numbers are :{npnum}")

def Divide(): # Division
  num=get_numbers()

  result=num[0]

  for i in range(1, len(num)):
      if num[i] == 0:
          print("Error: Division by zero!")
          break
      result = np.divide(result, num[i])
  else:        # only runs if no break
      print(f"The division of numbers is: {result}")

def Power():
  num=get_numbers()
  result=num[0]
  for i in range(1,len(num)):
      result=result**num[i]
  pow=result
  print(f"The Power calculation of numbers are :{pow}")


def Modulus():
  num=get_numbers()

  result=num[0]
  for i in range(1,len(num)):
    if num[i]==0:
      print("Error: Modulus by zero is not allowed!")
      return

    result=result % num[i]


  mod=result
  print(f"The modulus of  numbers are :{mod}")


while(True):
  print("----CONSOLE-BASED SMART CALCULATOR----")
  print("+ Addition")
  print("* Multiplication")
  print("- Subtraction")
  print("/ Division")
  print("** Power")
  print("%  Modulus")
  print("q Quit")

  o=input("Enter operation: ")

  if o=="q":
    print("Exiting. Thankyou!")
    break

  elif o=="+":
    Addition()

  elif o=="-" :
    Subtraction()

  elif o=="*" : # Product logic
    Product()

  elif o=="/": # Division logic
    Divide()

  elif o=="**": # Power Logic
    Power()

  elif o=="%": # Modulus Logic
    Modulus()

  else:
    print("Enter valid operator!")
```
